```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 05_cnn_ppt_32
# make filters

filters = np.zeros(shape=(5, 5, 3, 2), dtype = np.float32) # shape = (h, w, depth, # of filter)
filters[:, 2, :, 0] = 1 # vertical line
filters[2, :, :, 1] = 1 # horizontal line
# filters

print(filters[:, :, 0, 0]) # vertical line filter의 한 층만 print
plt.imshow(filters[:, :, :, 0])
plt.show()

print(filters[:, :, 0, 1]) # horizontal line filter의 한 층만 print
plt.imshow(filters[:, :, :, 1])
plt.show()
```

```python
# 05_cnn_ppt_35
# colormap 적용하기, colorbar, title 출력하기

filters = np.zeros(shape=(7, 7, 3, 2), dtype=np.float32)
filters[:, 3, :, 0] = 1 # vertical line, filter 0
filters[3, :, :, 1] = 1 # horizontal line, filter 1

f_0 = filters[:, :, :, 0] # vertical RGB
f_1 = filters[:, :, :, 1] # horizontal RGB

f_0_1 = filters[:, :, 0, 0] # vertical depth 1
f_1_1 = filters[:, :, 0, 1] # horizontal depth 1

# 흑백이미지 - viridis
plt.imshow(f_1_1) 
plt.colorbar() # image 옆에 colorbar 표시
plt.title('horizontal line filter')
plt.show()

# 흑백이미지 - gray
plt.imshow(f_1_1, cmap='gray')
plt.colorbar()
plt.title('gray horizontal line filter')
plt.show()

# 흑백이미지 - binary
plt.imshow(f_1_1, cmap='binary')
plt.colorbar()
plt.title('binary horizontal line filter')
plt.show()

# RGB - vertical
plt.imshow(f_0) 
plt.colorbar()
plt.title('vertical line filter')
plt.show()

# RGB - horizontal
plt.imshow(f_1)
plt.colorbar()
plt.title('horizontal line filter')
plt.show()
```
